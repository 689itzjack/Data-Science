# **OPERACIONES CON COLUMNAS**

**1. CREAR UNA COLUMNA NUEVA EN BASE A OTRA, MULTIPLICANDO EL RESULTADO DE LA COLUMNA A COPIIAR**

```python
df_mutations['PWM_ref_mult'] = df_mutations['PWM_ref'].apply(lambda x: x * 2)
# looking on the relevant columns, to check ourselves:
df_mutations[['PWM_ref', 'PWM_ref_mult']].head()
```

EN EL CODIGO DE ARRIBA CREAMOS UNA NUEVA COLUMNA CON EL NOMBRE 'PWM_ref_mult' Y LE METIOMOS A CADA CASILLA DE ELLA LA DATA QUE TIENE LA COLUMNA (TAMBN CASILLA POR CASILLA) 'PWM_ref' MULTIPLICANDOLA POR 2. DESPUES IMPRIMIMOS LAS PRIMERAS CASILLAS POR DEFAULT DE ESAS 2 COLUMNAS. 

***NOTA: LA FUNCION LAMBDA ES COMO UN ITERATOR, QUE PASA POR CADA UNA DE LAS CASILLAS DE LA COLUMNA QUE LA ACTIVA APLICANDO EN DICHA COLUMNA UNA OPERACION PARA GUARDARLA EN LA COLUMNA NUEVA EL RESULTADO DE LA OPERACION.***

**2. RESTAR DOS TABLAS PONIENDO EL RESULTADO EN UNA COLUMNA NUEVA**

```python
df_mutations['diff_ref_alt'] = df_mutations['PWM_ref'] - df_mutations['PWM_alt'] # RESTANDO LAS COLUMNAS 'PWM_ref'-'PWM_alt'
df_mutations[['PWM_ref', 'PWM_alt', 'diff_ref_alt']].head()
```

EN EL CODIGO DE  ARRIBA RESTANDO LAS COLUMNAS 'PWM_ref' - 'PWM_alt' Y PONIENDO EL RESULTADO EN LA COLUMNA NUEVA diff_ref_alt.
DESPES IMPRIMIMOS DICHAS COLUMNAS. 

LA RESTA SE EFECTUA CADA CASILLA DE UNA COLUMNA CON CADA CASILLA DE LA OTRA COLUMNA.   

**3. DIVISION ENTRE 2 COLUMNAS PONIENDO EL RESULTADO EN UNA NUEVA. HAY VECES QUE LA LLAMAN EL YAJAS ENTRE 2 COLUMNAS.**

```python
df_mutations['ratio_ref_alt'] = df_mutations['PWM_ref'] / df_mutations['PWM_alt']
df_mutations.head(10)
```

EL CODIGO DE ARRIBA ES LA MISMA EXPLICACION QUE LAS OPERACIONES DE ARRIBA PERO LA OPERACION ES LA DIVISION

**Part 5:** Count how many rows have ratio_ref_alt > 1 and how many have ratio_ref_alt <=1. What about the  rest of the rows?

```python
# Fetching rows with ratio > 1:
df_greater_1_ratio = df_mutations.loc[df_mutations['ratio_ref_alt'] > 1, :] # df_greater_1_ratio CONTIENE LAS HILERAS QUE CUPLIERON LA CONDICION
num_greater = len(df_greater_1_ratio)
num_greater
```

el codigo de arriba quiere decir:
df_mutations.loc[df_mutations['ratio_ref_alt'] > 1, :] TOMAMOS TODAS LA HILERAS ' : ' Y SOLO TOMAMOS AQUELLAS QUE SU 'ratio_ref_alt' ES MAYOR A 1.
DESPUES IMPRIMIMOS CUANTAS HILERAS CUMPLIERON CON DICHA CONDICION.

```python
# Fetching rows with ratio <= 1:
df_less_1_ratio = df_mutations.loc[df_mutations['ratio_ref_alt'] <= 1, :] # df_less_1_ratio CONTIENE LAS HILERAS QUE CUPLIERON LA CONDICION
num_less = len(df_less_1_ratio)
num_less
```

EL CODIGO DE ARRIBA ES LO MISMO QUE EL ANTERIOR PERO CON LA CONDICION MENOR O IGUAL A 1.

# **COMPROBANDO SI LAS TABLAS DE LA INFORMACION ADQUIRIDA CUMPLIERON CON LOS REQUISITOS**

```python
# assert(num_greater + num_less == len(df_mutations))
print(len(df_mutations) - num_greater - num_less) # RESTAMOS EL TOTAL DE TODAS LAS HILERAS DE LA TABLA Y LES RESTAMOS LOS RESULTADO DE LAS TABLAS OBTENIDAS
```

PODEMOS VER QUE LA RESTA DE ARRIBA NO ES 0, OSEA QUE TAL VEZ HAY HILERAS DE LA MISMA COLUMNA QUE CONTIENEN OTRO TIPO DE DATO

```python
df_mutations
```

en el indice 4 HAY UN VALOR NAN DE LA COLUMNA 'ratio_ref_alt' POR ESO EL RESULTADO NO DA 0, ENTONCES VAMOS A ADQUIRIR TODAS LAA HILERAS QUE CONTIENEN NAN DE LA COLUMNA 'ratio_ref_alt'.

```python
# Fetching rows having a NaN value (more about that - below)
num_nan = len(df_mutations.loc[df_mutations['ratio_ref_alt'].isnull(), :]) # ':' SIGNIFICA TODAS LAS HILERAS DE LA TABLA, ESTE COMM OBTIENE LAS HIELRAS NAN
num_nan
```

```python
# Sanity check: cheking that we covered all of the cases, and found all of the rows
assert(num_greater + num_less + num_nan == len(df_mutations)) # SE CAE EL PROGRAMA SI EL ASSERT NO PASA DEACUERDO ALA CONDICION
print(num_greater + num_less + num_nan, len(df_mutations))
```

# **RETOS DEL EJERCICIO**

**Part 6 (i):** Count how many rows have ratio_ref_alt > 1 and the "Group" value is "Positive": 

```python
import numpy as np 
np.unique(df_mutations['Group'].values) # CON ESTA FUNCION PODEMOS VER LOS VALORES QUE TIENE LA COLUMNA Group DE LA TABLA df_mutations 
#POR MEDIO DE LA FUNCION np.unique()
```

```python
a = (df_mutations.loc[:, 'ratio_ref_alt'] > 1) & (df_mutations.loc[:, 'Group'] == 'Positive')
# print(a)
print(a.sum()) # how many are True
```

la variable a CONTIENE UNA TABLA QUE CUMPLE CON LAS SIGUIENTES CONDICIONES:
1. TODAS LAS HILERAS LAS CUALES LOS ELEMENTOS DE LA COLUMNA 'ratio_ref_alt' SON MAYORES A 1.
2. TODOS LOS ELEMENTOS DE LAS HILERAS Y LA COLUMNA Group SEAN POSITIVOS

AL FINAL QUEREMOS VER CUANTOS ELEMENTOS SUMAN LOS QUE CUMPLEN CON LA CONDICION DE AMBAS COLUMNAS


```python
b = (df_mutations.loc[:, 'ratio_ref_alt'] > 1) & (df_mutations.loc[:, 'Group'] == 'Negative')
print(b.sum())
```

COMO ARRIBA PERO CON ELEMENTOS DE LA COLUMNA GROUP NEGATIVOS

```python
c = (df_mutations.loc[:, 'ratio_ref_alt'] <= 1) & (df_mutations.loc[:, 'Group'] == 'Positive')
print(c.sum())
```

```python
d = (df_mutations.loc[:, 'ratio_ref_alt'] <= 1) & (df_mutations.loc[:, 'Group'] == 'Negative')
print(d.sum())
```

IGUAL QUE ARRIBA PERO CON LA CONDICION <= 1

**ENEL CODIGO DE ABAJO OBSERVAMOS SI CUBRIMOS TODAS LAS CONDICIONES QUE DIJIMOS ANTERIORMENTE HACIENDO UN TEST**

```python
# Sanity check: check that we covered all cases:
all_cases = [a.sum(), b.sum(), c.sum(), d.sum(), num_nan] # AQUI TENEMOS LA SUMA DE TODAS LAS CONDICIONES INCLUYENDO LAS NAN
assert(sum(all_cases) == len(df_mutations)) # HACEMOS LA REVISION SI EL TEST PASA O NO
print('all_cases: ', all_cases)
print('sum(all_cases): ', sum(all_cases))
print('No. rows of original dataframe: ', len(df_mutations))
```
