# **COMO ADHERIR UNA COLUMNA NUEVA ALA DATA FRAME, OSEA LA TABLA**

```python
df_students['EDAD+5'] = 0
df_students
```

EN ESTE COMANDO INICIAMOS UNA COLUMNA NUEVA CON UN VALOR DEFAULTIVO 0 (PODEMOS TAMBN INICIARLO CON UN VALOR DIFULTIVO DIFERENTE)

```python
df_students['EDAD+5'] = df_students['EDAD'] + 5
df_students
```

**OTRA FORMA DE CREAR UN COLUMNA Y AGREGARLE DAATOS EN BASE A OTRA COLUMNA, ES HACER UNA FUNCION QUE SE ACTIVE AL MOMENTO DE METER LOS DATOS A LA COLUMNA**




```python
def add_Age_10(x):
  return x + 10

df_students['EDAD+10'] = df_students['EDAD'].apply(lambda x: add_Age_10(x))
df_students
```

EN EL COMANDO ANTERIOR CREAMOS LA COLUMNA EDAD+10 Y A LA COLUMNA EDAD LE APLICAMOS LA FUNCION "add_Age_10"
Y POR MEDIO DEL ITERATOR "lambda x", cada cuadro de informacion x (de la columna EDAD)le aplicamos la funcion add_Age_10 (agregando 10). EN OTRAS PALABRAS LA COL EDAD+10 ES LO QUE NOS DA LA FUNCION ACTIVADA EN LA COLUMNA EDAD.

```python
df_students['EDAD+7'] = df_students['EDAD'].apply(lambda x: x + 7)
df_students
```

TAMBIEN PODEMOS HACER LO CON UN ITERATOR SIN AYUDA DE LA FUNCION COMO EL COMMANDO DE ARRIBA

# **COMO GUARDAR EL DATA FRAME (LA TABLA) EN UN ARCHIVO .CSV O BINARIO**

```python
df_students
```

```python
null_mask = df_students['OCUPACION'].isnull()
not_null_mask = ~null_mask
print(null_mask)
print(not_null_mask)
df_students_new = df_students[not_null_mask]
df_students_new
```

EN EL COMANDO DE ARRIBA DEFINIMOS LA variable not_null_mask QUE ALMACENA UNA TABLA EN LA CUAL NO INCLUYE LOS ESTUDIANTES QUE NO TIENEN OCUPACION. DE AHI DEFINIMOS LA VARIABLE df_students_new QUE CONTIENE LA TABLA  sin los ESTUDIANTES QUE NO TIENEN OCUPACION

```python
df_students_new.to_csv('df_students_new.csv') #CREATE A NEW FILE NAMED df_students_new.csv
```

OTRO MODO MUY USADO ES GUARDAR LA TABLA EN UN FOORMATO BINARIO, SE GUARDA EN ESTE FORMATO CUANDO TENEMOS DEMASIADA INFORMACION. EL FORMATO DE ESTE TIPO DE ARCHIVO ES pickle (.pkl)
